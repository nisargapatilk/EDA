```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/ADVERK AI SHREY/datasets/train.csv')

# Display the first few rows of the dataset
print(data.head())

# Summary statistics
print(data.describe())

# Check for missing values
print(data.isnull().sum())
```

```python
data.info()
```

```python
data.isnull().sum()
#checking for null values
```

```python
#pre-process the data
for data in [data]:
    data["Age"].fillna(value=data["Age"].mean(), inplace=True)
    data["Fare"].fillna(value=data["Fare"].median(), inplace=True)
    data["Embarked"].fillna(value=data["Embarked"].mode()[0], inplace=True)
```

```python
#filling the null values by pre-processing as mentioned
```

```python
data.isnull().sum()
```

```python
data
#it is final DataFrame
```

```python
# Visualizations

# 1. Survival count
sns.countplot(x='Survived', data=data)
plt.title('Survival Count')
plt.show()
```

```python
# 2. Age distribution
sns.displot(data['Age'],kde=True)
plt.title('Age Distribution')

```

```python
# 3. Fare distribution
sns.displot(data['Fare'], bins=10, kde=True)
plt.title('Fare Distribution')
plt.show()

```

```python
# 4. Gender distribution
sns.countplot(x='Sex', data=data)
plt.title('Gender Distribution')
plt.show()

```

```python
# 5. Survival by Gender
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Survival by Gender')
plt.show()
```

```python
# 6. Survival by Pclass
sns.countplot(x='Survived', hue='Pclass', data=data)
plt.title('Survival by Pclass')
plt.show()
```

```python
 #7. Boxplot for Age and Pclass
sns.boxplot(x='Pclass', y='Age', data=data)
plt.title('Boxplot for Age and Pclass')
plt.show()
```
